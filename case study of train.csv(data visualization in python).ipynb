# Classes with regard to Age

```python
# now plotting  number of people  in different classes with regard to their age
#In statistics, kernel density estimation (KDE) is a non-parametric way to estimate the probability density 
#function (PDF) of a random variable. This function uses Gaussian kernels and includes automatic bandwidth determination.
titanic[titanic.Pclass==1]['Age'].plot(kind = 'kde',figsize = (10,5))
titanic[titanic.Pclass==2]['Age'].plot(kind = 'kde',figsize = (10,5))
titanic[titanic.Pclass==3]['Age'].plot(kind = 'kde',figsize = (10,5))
plt.title('classes with regard to age')
plt.legend(['1st','2nd','3rd'])
plt.show()
```

```python
# using 'for' loop for itirating through different class with ragards to the age

for i in [1,2,3]:
    titanic[titanic.Pclass== i]['Age'].plot(kind = 'kde',figsize = (10,5))
    plt.legend(['1st','2nd','3rd'])
    plt.title('classes with regard to age')
    
    
    
```

The  figure 1, shows that the people of age 25-55 have higher density whereas people above and below that age has 
comparatively lower density.Therefore, people of age 25-55 are more in Pclass 1.

The  figure 2, shows that the people of age 15-45 have higher density whereas people above and below that age has 
comparatively lower density.There are also few number ofchildren in that class .Therefore, people of age 15-45 are more 
in Pclass 2.

The  figure 3, shows that the people of age 10-35 have higher density whereas people above and below that age has 
comparatively lower density.There are also few number of children in that class .Therefore, people of age 10-35 are more 
in Pclass 3.

# Ploting data by defining Function

```python
def plot_bar_chart(feature):
    survived =titanic[titanic['Survived']==1 ][feature].value_counts()
    dead = titanic[titanic['Survived']==0][feature].value_counts()
    df = pd.DataFrame([survived,dead])
    df.index= ['Survived','Dead']
    df.plot(kind ='bar', figsize=(10,5),stacked=True)
```

```python
plot_bar_chart('Sex')
```

```python
plot_bar_chart('Embarked')
```

```python
plot_bar_chart('SibSp')
```
